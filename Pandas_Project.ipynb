## Merging & Concatenation

```python
import pandas as pd

var1 = {
    'A' : [1,2,3,4],
    'B' : [5,6,7,8]
}

df1 = pd.DataFrame(var1 , index = [1,2,3,4])

var2 = {
    'B':[5,6,7,9],
    'C':[9,10,11,12]
}

df2 = pd.DataFrame(var2 , index = [1,2,3,4])

pd.merge(df1 , df2 , how = "outer" , indicator = True)
```

```python
var1 = {
    'A' : [1,3,5,7],
    'B' : [2,4,6,8]
}

df1 = pd.DataFrame(var1)

var2 = {
    'A' : [5,6,9,6],
    'B' : [1,3,5,7]
}

df2 = pd.DataFrame(var2)

pd.merge(df1 , df2 , right_index = True , left_index = True , suffixes = ["id" , "name"])
```

```python
sr_1 = pd.Series([1,2,3,4])
sr_2 = pd.Series([5,6,7,None])
pd.concat([sr_1 , sr_2] , axis = 1 , keys = ["x_1" , "x_2"])
```

## Group by

```python
import pandas as pd

var = {
    "Section-wise" : ["a" , "a" , "a" , "a" , "b" , "b" , "b" , "c"],
    "A" : [23 , 46 , 54 , 39 , 67 , 87 , 45 , 13],
    "B" : [32 , 43 , 67 , 87 , 12 , 43 , 89 , 33]
}

df = pd.DataFrame(var)
df
```

```python
x_new = df.groupby("Section-wise")

for x,y in x_new:
  print(x)
  print(y)
  print()
```

```python
x_new.get_group("a")
```

```python
x_new.get_group("b")
```

```python
import pandas as pd

A = pd.DataFrame({"x1": [1,2,3,4] , "x2" : [5,6,7,8]} , index = ["a" , "b" , "c" , "d"])
B = pd.DataFrame({"x1" : [1,2,3] , "x2" : [4,5,6]} , index = ["a" , "b" , "c"])

A.join(B , how = "outer" , lsuffix = "id" , rsuffix = "name")
```

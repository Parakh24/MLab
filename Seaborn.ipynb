# Lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = {
    "x_1": [2,3,21,32,4,76,87],
    "x_2": [12,12,32,43,56,87,9] 
}

df = pd.DataFrame(dataset , index = [1,2,3,4,5,6,7])  
sns.lineplot(x="x_1", y="x_2" , data = dataset , marker = 'o' , mec = "red")
df 
```

```python
y_1 = sns.load_dataset("penguins").head(40) 
y_1
```

```python
sns.lineplot(x = "bill_length_mm" , y = "bill_depth_mm" , data = y_1 , hue = "sex" , style = "sex" , palette = "Accent" , markers = ["o" , ">"] , legend = "full")
plt.show()
```

# Histogram Plot

```python
var = sns.load_dataset("penguins").head(40) 
var
```

```python
sns.displot(var["flipper_length_mm"] , bins = [170,175,180,185,190,195,200,205,210,215] , kde = True , rug = True , color = 'red') 
plt.show()
```

```python
sns.displot(var["flipper_length_mm"]) 
plt.show() 
```

# Bar Plot

```python
var = sns.load_dataset("penguins")
var
```

```python
import matplotlib.pyplot as plt
```

```python
order_1 = ["Dream" , "Torgersen" , "Biscoe"] 
sns.barplot(x = "island" ,y = "bill_length_mm" , data = var , hue = "sex" , 
            order = order_1 , hue_order = ["Female" , "Male"] , palette = "icefire" ,
            saturation = 10 , errcolor = "green" , errwidth = 5 , capsize = 0.7)  
plt.show()   
```

# Scatter Plot

```python
var = sns.load_dataset("penguins").head(20) 
var
```

```python
sns.scatterplot(x = "bill_length_mm" , y = "bill_depth_mm" , data = var , hue = "sex" , style = "sex" , palette = "winter_r" , size = "sex" , sizes = (50,100) ,
                hue_order = ["Female" , "Male"] , alpha = 0.8) 
```

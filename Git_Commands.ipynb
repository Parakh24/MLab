# Git init

```
The git init command creates a new repository or reinitiates an existing one.When you run the
command, your current directory becomes a Git repository, allowing you to track its changes.


# Git Clone

```
To Copy a git repository from one location to another, we use the git clone command.It
usually copies an existing repository, including logs and versions, from remote servers
like git and github to your local machine.
```

**Syntax**
```git clone <copied URL>```



# Git Status

```
The git status command shows us details about:


*   modified files (files that are changed but not staged).
*   untracked files (files that Git is not tracking).
*   staged file (files that are staged and ready to be committed).
```

**Syntax**
```git status```



# Git add(for staging)

```

**Syntax** <br><br>
```git add . : stages the changes made to file```
  
    
```git add <filename> : stages the changes made to a particular file```

# Git Push


```
The Git Push command syncs your remote repository with the local
repository.Once you run this command, the remote repository will
reflect all the changes you committed locally.
```

**Syntax** <br>
```git push <remote> <branch>``` <br><br>
```<remote>: alias to the remote repository```<br><br>
```<branch>: name of the branch you want to push```


<br>

**Variations** <br>
```git push ---all origin: pushes all your local branches to your remote repositories```<br><br>

```git push --force origin main: This command overwrites the main branch on the remote repository with your local main branch. Be cautious while using it, as the remote main branch might lose all its history.```



# Git Pull

```The git pull command fetches and merges the changes in the remote repository with those in the local repository.```
<br><br>

```Git Fetch: retrieves all the changes made to the remote repository, including new commits, tags, and branches.```<br><br>
```Git Merge executes, integrating those changes to the current local branch.```

**Syntax**<br>

```git pull origin feature-branch: The above command fetches data from the feature_branch on the remote repository and commits it to your current local branch```

# Git Reset

```git reset: moves back to a previous commit but keeps current changes staged, allowing you to re-commit them later when needed.```
<br><br>

```git reset --soft HEAD~1: moves back to a previous commit and removes the current commit changes from the staging area.```
<br><br>
```git reset --mixed HEAD~1: moves back to a previous commit and removes the current commit changes from the stage area.```
<br><br>
```git reset --hard HEAD~1: This will point to a previous commit while removing current commit updates from the working directory and staging area.```

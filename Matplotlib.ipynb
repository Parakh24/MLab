### Mode

Value with the maximum frequency

```python
! pip install scipy
```

```python
from scipy import stats
value = [1,2,2,2,2,3,3,4,8,2,4,7,6,5,5,3,9] 
mode_salary = stats.mode(value)     
print(mode_salary[0])   # It counts the value of maximum frequency
print(mode_salary[1])   # It counts how many times the value has occured
freq = value.count(mode_salary[0])  # It also counts how many times the value has occured
freq1 = value.count(mode_salary[1])  # It counts the frequency of the number of counts of value 
print(freq)  
print(freq1)                                                                                                                
```

Measure of Dispersion

Range: Difference between maximum and minimum value

```python
value = [23,43,54,65,67,78,89]
range_value = np.max(value) - np.min(value)
print(range_value) 
```

```python
value = [10,11,12,13,14,15]
range_value = np.max(value) - np.min(value) 
print(range_value) 
```

```python
import numpy as np
```

```python
#Variance = how much the values differ in the list from the mean  
value = [1,2,3,4,5]   
mean_value =  np.mean(value)                           
variance_value = np.var(value)
print('Mean value is: ', mean_value)
print('Variance value is : ', variance_value)                         
```

```python
#Standard Deviation is square root of variance how far datasets from the mean 
std_val = np.std(value)  
print(std_val)               
```

Inter Quarentile Range (IQR)

```python
salary = [2343,2342,2333,7552,3456,1234]
salary.sort() 
Q1 = np.percentile(salary,25)
Q3 = np.percentile(salary,75)
print(salary)
print(Q1)
print(Q3) 
IQR = Q3-Q1
print(IQR)  
```

```python
min = Q1-1.5*IQR
max =  Q3+1.5*IQR
print(min)
print(max)        
```
